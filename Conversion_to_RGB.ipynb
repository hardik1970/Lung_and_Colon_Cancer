```python
import os
import csv

def create_label_csv(input_folder, output_csv_path, label):
    # Get a list of all files in the input folder
    file_list = os.listdir(input_folder)

    # Create or open the CSV file for writing
    with open(output_csv_path, 'w', newline='') as csvfile:
        csv_writer = csv.writer(csvfile)

        # Write the header row (filename, label)
        csv_writer.writerow(['filename', 'label'])

        # Write each image filename and its corresponding label to the CSV file
        for filename in file_list:
            csv_writer.writerow([filename, label])

# Replace 'input_folder_path' and 'output_csv_path' with your actual paths
input_folder_path = '/home/student/Downloads/SEMINAR/LC25000/lung_colon_image_set/colon_kaggle/colon_aca_rgb1'
output_csv_path = '/home/student/Desktop/colon_aca_label.csv'
label_value = 0

create_label_csv(input_folder_path, output_csv_path, label_value)
```

```python
import cv2
import os

# Specify the path to the directory containing your JPEG images
input_directory = '/home/student/Downloads/SEMINAR/LC25000/lung_colon_image_set/lung_image_sets/lung_scc'

# Specify the output directory where you want to save the converted images
output_directory = '/home/student/Downloads/SEMINAR/LC25000/lung_colon_image_set/lung_rgb/lung_scc_rgb'

# Ensure the output directory exists, create it if necessary
os.makedirs(output_directory, exist_ok=True)

# Loop through each JPEG image in the input directory
for filename in os.listdir(input_directory):
    if filename.lower().endswith('.jpg') or filename.lower().endswith('.jpeg'):
        # Read the image in BGR format
        bgr_image = cv2.imread(os.path.join(input_directory, filename))

        # Convert BGR to RGB
        rgb_image = cv2.cvtColor(bgr_image, cv2.COLOR_BGR2RGB)

        # Save the converted image to the output directory
        output_path = os.path.join(output_directory, filename)
        cv2.imwrite(output_path, rgb_image)

print("Conversion completed.")
```

```python
import os
import shutil

def process_folders(root_folder):
    # Iterate through all folders and subfolders
    for foldername, subfolders, filenames in os.walk(root_folder):
        # Delete files ending with 'CEO.txt' or 'CSR.txt'
        for filename in filenames:
            if filename.endswith(('CEO.txt', 'CSR.txt')):
                file_path = os.path.join(foldername, filename)
                os.remove(file_path)
                print(f"Deleted: {file_path}")

        # Create 'Annual_report' folder for remaining files
        remaining_files = [f for f in filenames if not f.endswith(('CEO.txt', 'CSR.txt'))]
        if remaining_files:
            annual_report_folder = os.path.join(foldername, 'Annual_report')
            os.makedirs(annual_report_folder, exist_ok=True)
            for file in remaining_files:
                source_path = os.path.join(foldername, file)
                dest_path = os.path.join(annual_report_folder, file)
                shutil.move(source_path, dest_path)
                print(f"Moved to Annual_report: {source_path} -> {dest_path}")

# Replace 'your_root_folder' with the actual root folder path
root_folder_path = '/home/student/Desktop/Task_3/Team 6_1'
process_folders(root_folder_path)
```

```python
import os
import shutil

def move_images(source_folder, destination_folder, num_images):
    # Ensure destination folder exists, if not, create it
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)

    # Get list of files in source folder
    files = os.listdir(source_folder)

    # Filter out only image files (you may adjust this as needed)
    image_files = [file for file in files if file.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp'))]

    # Limit the number of images to move, if specified
    if num_images > 0:
        image_files = image_files[:num_images]

    # Move each image file from source folder to destination folder
    for file in image_files:
        source_path = os.path.join(source_folder, file)
        destination_path = os.path.join(destination_folder, file)
        shutil.move(source_path, destination_path)
        print(f"Moved {file} to {destination_folder}")

# Replace 'source_folder' and 'destination_folder' with actual paths
source_folder = '/home/student/Downloads/SEMINAR/LC25000/lung_colon_image_set/lung_rgb/lung_scc_rgb'
destination_folder = '/home/student/Downloads/SEMINAR/LC25000/lung_colon_image_set/lung_kaggle/Test/lung_scc'
num_images = 1000  # Number of images to move, set to 0 to move all images

move_images(source_folder, destination_folder, num_images)
```
